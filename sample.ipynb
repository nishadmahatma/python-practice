## Comparison with lists

```python
import csv

with open('dataset.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

If we wanted for example for the third member of the experiment the value of the second value, we would need to do:

```python
data[2][1]
```

However, if we want the second value for all, we cannot do it. Let's use NumPy

```python
import numpy as np
```

```python
numpy_arr = np.array(data)
```

```python
numpy_arr[2,1]  # Notice we access like "matrices".
```

Get a column

```python
numpy_arr[:, 1]  # Returns a numpy array of the second column
```

Get a row

```python
numpy_arr[2,:]
```

## View vs copy

What happens if we change a value in a slice?

```python
a = numpy_arr[2]
```

```python
a
```

```python
a[3] = 5
```

```python
a
```

Notice now the fourth value is 5

```python
numpy_arr[2,3]
```

And now the value was changed in the original array. Slices are "views" of the original numpy array. If we want this to not happen:

```python
a = np.copy(numpy_arr)[2]
a[3] = 10
a
```

```python
numpy_arr[2,3]
```

Notice it is not 10!

## Partial slicing

If we don't want the full column or row, we can do a partial slice

```python
# Get me from the second row, from the third to the seventh value not inclusive
numpy_arr[1, 2:6]
```

Notice the : to delimiter the start and end of the slice. In general the initial value is included and the end value is not included; everything 0 based

```python
# From second to second to last row, the first two columns
numpy_arr[1:-1, :2]
```
